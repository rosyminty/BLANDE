## Splitting

```python
g_pr_d = g_pr[g_pr['check'] != 2]
g_pr_d
```

```python
len(g_pr_d)
```

```python
# 인덱싱 다시
g_pr_d.reset_index(drop=True, inplace=True)
```

```python
g_pr_d.loc[0, 'contents']
```

```python
g_pr_list = []
for idx in range(len(g_pr_d)) :
    cont = []
    if g_pr_d.loc[idx, 'check'] == 1 :
        cont.append(g_pr_d.loc[idx, 'contents'])
    else :
        for content in g_pr_d.loc[idx, 'contents'].split("', '") :
#             print(idx)
#             print(content)
#             print('*'*100)
            if '제네시스' in content :
                cont.append(content)
#                 print(content)
        print(len(cont))
    g_pr_list.append(cont)
len(g_pr_list)
```

```python
g_pr_list[4]
```

```python
g_pr_d['word_genesis'] = g_pr_list
g_pr_d
```

```python
# 중간저장
g_pr_d.to_csv('./data/genesis/g_pr.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```
