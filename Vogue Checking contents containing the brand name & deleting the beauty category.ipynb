```python
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
import time
```

```python
V_D_cc = pd.read_csv('./data/Vogue_Dior_ccc.csv' , sep='|', encoding='utf-8')
V_D_cc
```

### Dior

```python
# title에 '디올' 없는 행 중 contents에도 '디올' 없으면 check=2로 수정
```

```python
for i in range(len(V_D_cc)) :
    if V_D_cc['title_brand'][i] == 0 :
        if not '디올' in V_D_cc['contents'][i] :
            V_D_cc.iloc[i,4] = 2
            print(i)
    else : 
        pass
```

```python
'디올' in V_D_cc['contents'][66]
```

```python
V_D_cc
```

```python
# checking Beauty category
```

```python
Beauty = []
for i in range(len(V_D_cc)) :
    if V_D_cc['category'][i] == 'Beauty' :
        V_D_cc.iloc[i,4] = 2
        Beauty.append(i)
    else : 
        pass
print(Beauty)
```

```python
V_D_cc.iloc[Beauty]
```

```python
V_D_cc.to_csv('./data/Vogue_Dior_ccc.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```
