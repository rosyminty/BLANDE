### 크롤링 부재한 아이들 채우기

```python
# contents가 공백인 애들 다시 지정
cblank = []
for i in range(len(V_R)) :
    if V_R['contents'][i] == '[]' :    
        cblank.append(i)
print(cblank)
print(len(cblank))
```
