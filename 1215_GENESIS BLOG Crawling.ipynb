```python
title_list = []
category_list = []
link_list = []

articles = driver.find_elements_by_css_selector('ul.thumbnail-a4 > li')
    
for article in articles :
    
    category = article.find_element_by_css_selector('a > div > span').text
    category_list.append(category)
    print(category)
    
    title = article.find_element_by_css_selector('a > div > span').text
    title_list.append(title)
    print(title)

    link = article.find_element_by_css_selector('a').get_attribute('href')
    link_list.append(link)
    print(link)
    
    print('*'*100)
```

```python
print(len(title_list))
print(len(category_list))
print(len(link_list))
```

```python
HMGJ_GENESIS_BLOG = pd.DataFrame({
    'title' : title_list, 
    'category'  : category_list,
    'link'  : link_list
})
```
