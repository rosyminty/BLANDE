### Celine

```python
baseUrl = 'http://www.vogue.co.kr/page/'
subUrl = '/?s='
plusUrl = '셀린'
# url = baseUrl + urllib.parse.quote_plus(plusUrl)
# url
```

```python
vUrl = baseUrl+str(1)+subUrl+urllib.parse.quote_plus(plusUrl)
vUrl
```

```python
path = './driver/chromedriver.exe'
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(vUrl)
```

```python
title_list = []
date_list = []
link_list = []
contents = []

for i in range(22) :
    vUrl = baseUrl+str(i+1)+subUrl+urllib.parse.quote_plus(plusUrl)
    driver.get(vUrl)
    
    articles = driver.find_elements_by_css_selector('section#content > div > div > article')
#     len(articles)

    for article in articles :
        title = article.find_element_by_css_selector('div.fusion-post-wrapper > div.fusion-post-content-wrapper > div > h2 > a').text
        title_list.append(title)
        print(title)
    
        date = article.find_element_by_css_selector('div.fusion-post-wrapper > div.fusion-post-content-wrapper > div > p').text
        date_list.append(date)
        print(date)
    
        link = article.find_element_by_css_selector('div.fusion-post-wrapper > div.fusion-post-content-wrapper > div > h2 > a').get_attribute('href')
        link_list.append(link)
        print(link)
    
        print('-'*50)
    
#         print(vUrl)
        
```

```python
len(title_list)
```

```python
len(link_list)
```

```python
len(date_list)
```

```python
Vogue_SaintLaurent_df = pd.DataFrame({
    'title' : title_list,
    'date'  : date_list, 
    'link'  : link_list
})
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
```

```python
contents = []
for i in range(len(V_Ce)) :
    cUrl = V_Ce['link'][i]
#     print(cUrl)
    driver.get(cUrl)
    pghs = driver.find_elements_by_css_selector('div.post-content > p')
#     print(pghs)
    paragraphs = []
    for i in range(len(pghs)) :
        pgh = pghs[i].text
        paragraphs.append(pgh)
    contents.append(paragraphs)
    print(paragraphs)
    print('-'*100)
```

```python
len(contents)
```

```python
V_Ce['contents'] = contents
V_Ce
```
