```python
import pandas as pd
import numpy as np
```

## load files

```python
V_Je = pd.read_csv('./data/Vogue_JENNIE_df.csv' , sep='|', encoding='utf-8')
V_Ji = pd.read_csv('./data/Vogue_JISOO_df.csv' , sep='|', encoding='utf-8')
V_L = pd.read_csv('./data/Vogue_LISA_df.csv' , sep='|', encoding='utf-8')
V_R = pd.read_csv('./data/Vogue_ROSE_df.csv' , sep='|', encoding='utf-8')
```

## JENNIE

```python
v_je_d = V_Je[V_Je['check'] != 2]
v_je_d
```

```python
v_je_list = []
for i in range(len(v_je_d)):
    cont = []
    if v_je_d.iloc[i,8] == 1 :
        cont.append(v_je_d.iloc[i,6])
    else :
        for content in v_je_d.iloc[i, 6].split("', '") :
#             print(i)
#             print(content)
#             print('-'*50)
            if '제니' in content:
                cont.append(content)
#                 print(content)
        print(len(cont))
    v_je_list.append(cont)

len(v_je_list)
```

```python
print(v_je_list[3])
```

```python
v_je_d.loc[1,'check']
```

```python
v_je_d['word_jennie'] = v_je_list
v_je_d
```
