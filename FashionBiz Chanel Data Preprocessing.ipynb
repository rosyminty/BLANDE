## contents crawling

```python
path = './driver/chromedriver.exe'
```

```python
driver_Ch = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver_Ch
```

```python
contents = []
for i in range(len(F_Ch)) :
    cUrl = F_Ch['link'][i]
#     print(cUrl)
    driver_Ch.get(cUrl)
    time.sleep(0.5)
    pghs = driver_Ch.find_elements_by_css_selector('div.view_cont')
    if len(pghs) > 0 :
        content = pghs[0].text
        contents.append(content)
    else :
        content = ''
        contents.append(content)
    print('[',i,']' , content)
    print('-'*100)
```
