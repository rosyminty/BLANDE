```python
title_list = []
date_list = []
link_list = []

articles = driver.find_elements_by_css_selector('ul.thumbnail-b2 > li')
    
for article in articles :
    title = article.find_element_by_css_selector('div > strong.tit').text
    title_list.append(title)
    print(title)
    
    date = article.find_element_by_css_selector('div > strong.tit > span').text
    date_list.append(date)
    print(date)
    
    link = article.find_element_by_css_selector('a').get_attribute('href')
    link_list.append(link)
    print(link)
    
    print('*'*100)
```

```python
print(len(title_list))
print(len(date_list))
print(len(link_list))
```

```python
HMGJ_GENESIS_PR = pd.DataFrame({
    'title' : title_list, 
    'date'  : date_list,
    'link'  : link_list
})
```

```python
title_list = []
contents = []
tag_list = []

for i in range(len(g_pr)) :
    gUrl = g_pr.loc[i, 'link']
    driver.get(gUrl)
    
    time.sleep(0.5)
    
    print(i)
    
    title = driver.find_element_by_css_selector('div.tit_view > h1').text
    title_list.append(title)
    print(title)
    
    pghs = driver.find_elements_by_css_selector('p')
    paragraphs = []
    for i in range(len(pghs)) : 
        pgh = pghs[i].text
        paragraphs.append(pgh)
    contents.append(paragraphs)
    print(paragraphs)
    
    tags = driver.find_elements_by_css_selector('div.tag-area > a')
    tag_l = []
    for tag in tags :
        tag_l.append(tag.text)
    tag_list.append(tag_l)
    print(tag_l)
    print('*'*100)
```

```python
print(len(title_list))
print(len(contents))
print(len(tag_list))
```

```python
g_pr['title'] = title_list
g_pr['contents'] = contents
g_pr['tags'] = tag_list
g_pr
```

```python
g_pr.to_csv('./data/genesis/HMGJ_GENESIS_PR.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```
