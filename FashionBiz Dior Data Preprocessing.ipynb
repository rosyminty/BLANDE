### FOR loop for contents crawling

```python
contents = []
for i in range(len(F_D)) :
    cUrl = F_D['link'][i]
#     print(cUrl)
    driver.get(cUrl)
    time.sleep(0.5)
    pghs = driver.find_elements_by_css_selector('div.view_cont')
    if len(pghs) > 0 :
        content = pghs[0].text
        contents.append(content)
    else :
        content = ''
        contents.append(content)
    print('[',i,']' , content)
    print('-'*100)
```
