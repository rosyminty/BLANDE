### FOR loop for contents crawling

```python
contents = []
for i in range(len(F_D)) :
    cUrl = F_D['link'][i]
#     print(cUrl)
    driver.get(cUrl)
    time.sleep(0.5)
    pghs = driver.find_elements_by_css_selector('div.view_cont')
    if len(pghs) > 0 :
        content = pghs[0].text
        contents.append(content)
    else :
        content = ''
        contents.append(content)
    print('[',i,']' , content)
    print('-'*100)
```

### Category 가져오기

```python
# Dior
# https://www.fashionbiz.co.kr/article/search.asp?page=3&keyword=%B5%F0%BF%C3&keyword_option=&
```

```python
path = './driver/chromedriver.exe'
```

```python
baseUrl = 'https://www.fashionbiz.co.kr/article/search.asp?page='
plusUrl = '&keyword=%B5%F0%BF%C3&keyword_option=&'
```

```python
vUrl = baseUrl+str(1)+plusUrl
type(vUrl)
print(vUrl)
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(vUrl)
```

```python
categories = driver.find_elements_by_css_selector('div.txt_box > p.txt02 > a')
categories[0].text
```

```python
category_list = []

for i in range(3) :
    vUrl = baseUrl+str(i+1)+plusUrl
    driver.get(vUrl)
    time.sleep(0.5)
    
    categories = driver.find_elements_by_css_selector('div.txt_box > p.txt02 > a')
    for idx in range(len(categories)) :
        category = categories[idx].text
        category_list.append(category)
        print(i, idx, category)
```
