```python
import pandas as pd
import numpy as np
```

## Dior

```python
F_D = pd.read_csv('./data/FashionBiz_Dior_contents.csv' , sep='|', encoding='utf-8')
F_D
```

### 날짜 지난 거 빼기

```python
d = list(range(35,45))
d
```

```python
F_D.drop(d, inplace=True)
```

```python
F_D
```

```python
# 중간 저장
F_D.to_csv('./data/FashionBiz_Dior_contents.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```

### category 확인

```python
# category 종류
F_D['category'].unique()
```

```python
# category별 개수
F_D['category'].value_counts()
```
