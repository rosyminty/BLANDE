```python
from collections import Counter
```

```python
# 불용어 제외 리스트
noun_adj_list_stp= []
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if word not in stpwds_l:
            if tag in ['NNG','NNP','VA', 'VV', 'OL', 'UN']:
                noun_adj_list_stp.append(word)
```
