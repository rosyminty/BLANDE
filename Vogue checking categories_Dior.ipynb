### 크롤링 부재한 아이들만 따로

```python
from selenium import webdriver
import time
```

```python
path = './driver/chromedriver.exe'
```

```python
cUrl = V_D['link'][175]
cUrl
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
```

```python
driver.get(cUrl)
```

```python
c = driver.find_elements_by_css_selector('div.fusion-text > p')
c[2].text
```

```python
content = []
for i in blank :
    paragraph = []
    cUrl = V_D['link'][i]
    driver.get(cUrl)
#     print(cUrl)
    time.sleep(0.5)
    c = driver.find_elements_by_css_selector('div.fusion-text > p')
    for idx in range(len(c)) :
        p = c[idx].text
        paragraph.append(p)
#     V_D.iloc[i,6] = paragraph
    print(paragraph)
    print('-'*100)
    content.append(paragraph)
```

```python
len(content)
```

```python
V_D.iloc[blank,6] = content
V_D.iloc[blank]
```
