```python
import pandas as pd
import numpy as np
```

### Dior

```python
F_D = pd.read_csv('./data/FashionBiz_Dior_df.csv' , sep='|', encoding='utf-8')
F_D
```

```python
Dior = []
for i in range(len(F_D['title'])) :
    if '디올' in F_D['title'][i] :
        Dior.append(1)
        print(1)
    else :
        Dior.append(0)
        print(0)
```

```python
len(Dior)
```

```python
F_D['디올'] = Dior
F_D
```

```python
F_D.to_csv('./data/FashionBiz_Dior_df.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```

```python
F_D.to_excel('./data/FashionBiz_Dior_df.xlsx', index=False)
print('success')
```
