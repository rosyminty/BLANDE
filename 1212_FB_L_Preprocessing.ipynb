```python
FB_L['check'] = 0
```

```python
title_brand = []
for idx in range(len(FB_L)) : 
    if '알리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '엘리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '리사이클' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '요리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '멜리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '마리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 2
        title_brand.append(0)
    elif '리사' in FB_L['title'][idx] :
        FB_L.loc[idx, 'check'] = 1
        title_brand.append(1)
    else :
        title_brand.append(0)
```

### word_lisa 생성

```python
fb_l_d = FB_L[FB_L['check'] != 2]
fb_l_d
```

```python
fb_l_list = []
for i in range(len(fb_l_d)):
    cont = []
    if fb_l_d.iloc[i,8] == 1 :
        cont.append(fb_l_d.iloc[i, 6])
    else :
        for content in fb_l_d.iloc[i, 6].split("\n\n") :
#             print(i)
#             print(content)
#             print('-'*50)
            if '리사' in content:
                cont.append(content)
#                 print(content)
        print(len(cont))
    fb_l_list.append(cont)

len(fb_l_list)
```

```python
fb_l_d['word_lisa'] = fb_l_list
fb_l_d
```

```python
fb_l_d.to_csv('./data/FB_L.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```
