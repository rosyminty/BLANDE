```python
g_b_list = []
for idx in range(len(g_b)) :
    cont = []
    if g_b.loc[idx, 'title_brand'] == 1 :
        cont.append(g_b.loc[idx, 'contents'])
    else :
        for content in g_b.loc[idx, 'contents'].split("\', \'") :
#             print(idx)
#             print(content)
#             print('*'*100)
            if '제네시스' in content :
                cont.append(content)
#                 print(content)
        print(len(cont))
    g_b_list.append(cont)
len(g_b_list)
```
