## Kkma

```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
slsl = []
for line in magazine_sl['word_laurent']:
    slsl.append(line)
```

```python
type(slsl)
```

```python
for idx in range(len(slsl)) :
# for sentence in slsl :
    if '생로랑' in slsl[idx] :
        slsl[idx] = slsl[idx].replace('생로랑' , 'SAINT_LAURENT')
```

```python
for idx in range(len(slsl)) :
    if '이브 SAINT' in slsl[idx] :
        slsl[idx] = slsl[idx].replace('이브 SAINT' , 'YVES_SAINT')
```

```python
for idx in range(len(slsl)) :
    if '입SAINT' in slsl[idx] :
        slsl[idx] = slsl[idx].replace('입SAINT' , 'YVES_SAINT')
#         print(idx, slsl[idx])
```

```python
for idx in range(len(slsl)) :
    if '이브SAINT' in slsl[idx] :
        slsl[idx] = slsl[idx].replace('이브SAINT' , 'YVES_SAINT')
#         print(idx, slsl[idx])
```

```python
slsl
```

```python
# 리스트를 문자열로 합치기
text = ''.join(slsl)
# text
```

```python
# 각 문장별로 형태소 구분하기
sentences_tag = []
for i in range(len(slsl)) :
# for sentence in slsl:
    print(i)
    morph = kkma.pos(slsl[i])
    sentences_tag.append(morph)
#     print(morph)
#     print('-'*30)
```

```python
# 명사 혹은 형용사인 품사만 선별해 리스트에 담기
noun_adj_list = []
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['NNG','VA', 'VV', 'OL']:
            noun_adj_list.append(word)
```

```python
from collections import Counter
```

```python
# noun_adj_list
```

```python
# 별된 품사별 빈도수 계산 & 상위 빈도 50위 까지 출력
counts = Counter(noun_adj_list)
print(counts.most_common(50))
```

```python
# 명사만
noun_list = []
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['NNG', 'OL']:
            noun_list.append(word)
```

```python
noun_list
```

```python
counts_noun = Counter(noun_list)
print(counts_noun.most_common(50))
```
