### LISA

```python
# LISA
# https://www.fashionbiz.co.kr/article/search.asp?page=7&keyword=%B8%AE%BB%E7&keyword_option=&
```

```python
baseUrl = 'https://www.fashionbiz.co.kr/article/search.asp?page='

plusUrl = '&keyword=%B8%AE%BB%E7&keyword_option=&'
```

```python
vUrl = baseUrl+str(1)+plusUrl
type(vUrl)
print(vUrl)
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(vUrl)
```

```python
title_list = []
category_list = []
date_list = []
link_list = []

for i in range(7) :
    vUrl = baseUrl+str(i+1)+plusUrl
    driver.get(vUrl)
    time.sleep(0.5)
    
    articles = driver.find_elements_by_css_selector('div.comb_cont > ul > li')
    len(articles)

    for article in articles :
        title = article.find_element_by_css_selector('div.txt_box > h3 > a').text
        title_list.append(title)
        print(title)
    
        category = article.find_element_by_css_selector('div.txt_box > p.txt02 > a').text
        category_list.append(category)
        print(category)
        
        date = article.find_element_by_css_selector('div.txt_box > div').text
        date_list.append(date)
        print(date)
    
        link = article.find_element_by_css_selector('div.txt_box > h3 > a').get_attribute('href')
        link_list.append(link)
        print(link)
    
        print('-'*100)
        
```

```python
print(len(title_list))
print(len(category_list))
print(len(link_list))
print(len(date_list))
```

```python
FashionBiz_LISA_df = pd.DataFrame({
    'title' : title_list,
    'date'  : date_list, 
    'link'  : link_list,
    'category' : category_list
})
```

### contents crawling

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
```

```python
contents = []
for i in range(len(FashionBiz_LISA_df)) :
    cUrl = FashionBiz_LISA_df['link'][i]
#     print(cUrl)
    driver.get(cUrl)
    time.sleep(0.5)
    pghs = driver.find_elements_by_css_selector('div.view_cont')
    if len(pghs) > 0 :
        content = pghs[0].text
        contents.append(content)
    else :
        content = ''
        contents.append(content)
    print('[',i,']' , content)
    print('-'*100)
```

```python
FB_L = FashionBiz_LISA_df
```

```python
FB_L['contents'] = contents
```
