```python
baseUrl = 'https://www.fashionbiz.co.kr/article/search.asp?page='
# subUrl = '/?s='
plusUrl = '&keyword=%BB%FD%B7%CE%B6%FB&keyword_option=&'
# 이상한 함수 써서 '디올' 변환시키는 거가 안먹힘 ㅜ
# url = baseUrl + urllib.parse.quote_plus(plusUrl)
# url
```

```python
vUrl = baseUrl+str(1)+plusUrl
type(vUrl)
print(vUrl)
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(vUrl)
```

```python
# 첫 페이지만 테스트

# vUrl = baseUrl+str(i+1)+subUrl+urllib.parse.quote_plus(plusUrl)
# driver.get(vUrl)

title_list = []
date_list = []
link_list = []

articles = driver.find_elements_by_css_selector('div.comb_cont > ul > li')
#     len(articles)


for article in articles :
    title = article.find_element_by_css_selector('div.txt_box > h3 > a').text
    title_list.append(title)
    print(title)
    
    date = article.find_element_by_css_selector('div.txt_box > div').text
    date_list.append(date)
    print(date)
    
    link = article.find_element_by_css_selector('div.txt_box > h3 > a').get_attribute('href')
    link_list.append(link)
    print(link)
    
    print('-'*50)
```

```python
title_list = []
date_list = []
link_list = []

for i in range(2) :
    vUrl = baseUrl+str(i+1)+plusUrl
    driver.get(vUrl)
    
    articles = driver.find_elements_by_css_selector('div.comb_cont > ul > li')
    len(articles)

    for article in articles :
        title = article.find_element_by_css_selector('div.txt_box > h3 > a').text
        title_list.append(title)
        print(title)
    
        date = article.find_element_by_css_selector('div.txt_box > div').text
        date_list.append(date)
        print(date)
    
        link = article.find_element_by_css_selector('div.txt_box > h3 > a').get_attribute('href')
        link_list.append(link)
        print(link)
    
        print('-'*50)
    
#         print(vUrl)
        
```

```python
FashionBiz_SaintLaurent_df = pd.DataFrame({
    'title' : title_list,
    'date'  : date_list, 
    'link'  : link_list
})
```
