```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
```

```python
search_keyword = '디올'
url = 'http://www.vogue.co.kr/?s='+urllib.parse.quote_plus(search_keyword)
url
```

```python
webpage = requests.get(url)
webpage.text
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
soup
```

```python
soup.h2
```

```python
print(soup.h2.string)
```

```python
for child in soup.h2.children :
    print(child)
```

```python
soup.p
```

```python
for d in soup.div.children :
    print(d)
```

```python
print(soup.find_all('h2'))
```

```python
titles = soup.find_all('h2', {'class' : 'entry-title fusion-post-title'})
titles
```

```python
titles[0]
```

```python
titles[0].string
```

```python
titles[0].a.attrs['href']
```

```python
title_list = []
link_list = []
for i in titles : 
    print(i.string)
    print(i.a.attrs['href'])
    print()
    title_list.append(i.string)
    link_list.append(i.a.attrs['href'])
```

```python
soup.find_all('span','updated rich-snippet-hidden').attrs
```

```python
import pandas as pd
vogue_Dior_df = pd.DataFrame({
    'title' : title_list ,
    'link'  : link_list
})
```
