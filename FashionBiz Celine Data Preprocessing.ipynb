## contents crawling

```python
path = './driver/chromedriver.exe'
```

```python
driver_Ce = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver_Ce
```

```python
contents = []
for i in range(len(F_Ce)) :
    cUrl = F_Ce['link'][i]
#     print(cUrl)
    driver_Ce.get(cUrl)
    time.sleep(0.5)
    pghs = driver_Ce.find_elements_by_css_selector('div.view_cont')
    if len(pghs) > 0 :
        content = pghs[0].text
        contents.append(content)
    else :
        content = ''
        contents.append(content)
    print('[',i,']' , content)
    print('-'*100)
```
