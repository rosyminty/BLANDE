```python
import pandas as pd
import numpy as np
```

```python
FB_SL = pd.read_csv('./data/FashionBiz_SaintLaurent_contents.csv' , sep='|', encoding='utf-8')
FB_SL
```

```python
FB_SL.to_csv('./data/FB_SaintLaurent.csv' , sep='|', mode='w' , encoding='utf-8' , index=False)
print('success')
```
